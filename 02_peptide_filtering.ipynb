```python
import pandas as pd
```

```python
df = pd.read_csv('braun2025_peptide_rank.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
def set_detection_count(row, cols):
    count = 0
    for col in cols:
        if pd.notna(row[col]) and row[col] != '-' and row[col] != 0:
            count += 1
    return count
```

```python
# check overlap between 1106 and A431 keratinocyte cell line peptide rankings

cols_1106 = ['1106/Nil 1', '1106/Nil 2', '1106/Nil 3', '1106/IFN 1', '1106/IFN 2', '1106/IFN 3']
cols_A431 = ['A431_sC6/Rep 1', 'A431_sC6/Rep 2', 'A431_sC6/Rep 3']

df['1106_median_intensity'] = df[cols_1106].median(axis=1, skipna=True)
df['A431_median_intensity'] = df[cols_A431].median(axis=1, skipna=True)

df['1106_detection_count'] = df.apply(set_detection_count, axis=1, args=[cols_1106])
df['A431_detection_count'] = df.apply(set_detection_count, axis=1, args=[cols_A431])

ranked_1106 = df.sort_values(by=['1106_detection_count','NetMHCpan.4 \nScore'], ascending=[False, True]).reset_index(drop=True).reset_index()
ranked_1106.rename(columns={'index':'1106_rank'}, inplace=True)
ranked_1106['1106_rank'] += 1

ranked_A431 = df.sort_values(by=['A431_detection_count','NetMHCpan.4 \nScore'], ascending=[False, True]).reset_index(drop=True).reset_index()
ranked_A431.rename(columns={'index':'A431_rank'}, inplace=True)
ranked_A431['A431_rank'] += 1

ranked_1106 = ranked_1106[['Accession', '1106_rank', 'Peptide', 'ensembl ID']]
ranked_A431 = ranked_A431[['Accession', 'A431_rank', 'Peptide', 'ensembl ID']]

df = df.merge(ranked_1106, on=['Accession', 'Peptide', 'ensembl ID']).drop_duplicates('1106_rank')
df = df.merge(ranked_A431, on=['Accession', 'Peptide', 'ensembl ID']).drop_duplicates('A431_rank')
```

```python
df.drop_duplicates('Peptide', inplace=True)
df.shape
```

```python
ker_overlap = ranked_1106.head(250).merge(ranked_A431.head(250), on=['Peptide'])
ker_overlap.shape
```

```python
# get top 250 ranked peptides for each cell line (1106, A431, VMM1) and GTEx skin leg sun exposed
n_peptides = 16

small_df = df.drop(columns=['Peptide (with PTM)', 'NetMHCpan.4 \nScore', 'Gene stable ID version', 'Description', 'Accession', 'keratinocyte_rank', 'keratinocyte_detection_count', 'keratinocyte_median_intensity'], axis=1)
small_df.rename(columns={'melanocyte_rank':'VMM1_rank', 'melanocyte_median_intensity':'VMM1_median_intensity', 'melanocyte_detection_count':'VMM1_detection_count'}, inplace=True)

top_VMM1 = small_df.sort_values(by='VMM1_rank').head(n_peptides)[['Peptide']]
top_1106 = small_df.sort_values(by='1106_rank').head(n_peptides)[['Peptide']]
top_A431 = small_df.sort_values(by='A431_rank').head(n_peptides)[['Peptide']]
top_gtex = small_df.sort_values(by='gtex_rank').head(n_peptides)[['Peptide']]
```

```python
top_VMM1['selection_criteria'] = 'VMM1'
top_1106['selection_criteria'] = '1106'
top_A431['selection_criteria'] = 'A431'
top_gtex['selection_criteria'] = 'GTEx'
```

```python
top_VMM1.head()
```

```python
peptides = pd.concat([top_VMM1, top_1106, top_A431, top_gtex])
peptides_grouped = peptides.groupby('Peptide')['selection_criteria'].agg(lambda x: list(sorted(set(x)))).reset_index()

top_peptides = peptides_grouped.merge(small_df, how='left', on='Peptide')
```

```python
top_peptides.sort_values(by=['VMM1_rank', '1106_rank', 'A431_rank', 'gtex_rank'], ascending=True, inplace=True)
top_peptides.head(30)
```

```python
top_peptides.to_csv("ranked_peptides.csv", index=False)
```
